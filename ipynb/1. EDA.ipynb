```python
import pandas as pd
from os import path
import matplotlib.pyplot as plt
```

## Data-type
- delta: angle difference between the machine and the bus.
- omega: angle speed of the machine.
- Pm: __ of the machine.
- Eap: __ of the machine.
- Vmag: voltage magnituge of the bus.
- theta: voltage angle of the bus.
- E1: __ of the exciter.
- Efd: __ of the exciter.
- P3: __ of the govenor.
- Temperature: temperature of the relay.

```python
data_type = [
    "delta", "omega", "Pm", "Eap", "Vmag", "theta", "E1", "Efd", "P3",
    "Temperature"
]
```
